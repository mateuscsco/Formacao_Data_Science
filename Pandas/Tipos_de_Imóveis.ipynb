## Tipos de Imóveis

```python
import pandas as pd#importando o package do pandas
from google.colab import drive#importando o drive do google colab
drive.mount('/content/drive')#acessando o drive para utilizar arquivos do meu drive
```

```python
dados=pd.read_csv('drive/MyDrive/Colab Notebooks/Formação Data Science/Pandas/Dados/aluguel.csv', sep=";")#lendo arquivo com dados separados por ;
```

```python
dados.head(10)#exibindo 10 primeiras rows
```

```python
dados['Tipo']#exibindo apenas uma column em forma de Series(array)
```

```python
tipo_de_imovel = dados['Tipo']
type(tipo_de_imovel)#tipo Series é uma array unidimensional
```

```python
tipo_de_imovel.drop_duplicates()#Este método agrupa todos os tipos de imóveis existentes na série coletando apenas o primeiro item de cada tipo, e eliminando os outros.
```

```python
tipo_de_imovel.drop_duplicates(keep='last', inplace=False)#utilizando argumentos da função podemos determinar, por exemplo, o keep que pode manter, ao invés do primeiro de cada valor, o último.
```

```python
tipo_de_imovel.drop_duplicates(inplace=True)#já o inplace atribui o resultado da função à variável já durante a execução, caso igual a 'True', e não atribui, caso 'False'
tipo_de_imovel
```
