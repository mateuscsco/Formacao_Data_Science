## Imóveis Residenciais

```python
import pandas as pd#importando o pandas
from google.colab import drive #importando drive so google colab
drive.mount('/content/drive')#acessando google drive utilizando a função mount() do drive
```

```python
dados = pd.read_csv('drive/MyDrive/Colab Notebooks/Formação Data Science/Pandas/Dados/aluguel.csv', sep = ';')#importando csv
dados.head(10)
```

```python
residencial = list(dados['Tipo'].drop_duplicates()) #cria lista com a coluna Tipo, retirando valores em duplicidade
residencial
```

```python
residencial = ['Quitinete', #pegamos todos os tipos considerados residenciais e atribuímos à lista residencial
'Casa',
'Apartamento',
'Casa de Condomínio',
'Casa de Vila']
dados['Tipo'].isin(residencial).head(10)#verificamos quais valores em Tipo estão na lista denominada residencial, o que nos retorna uma Series com valores booleanos
```

```python
selecao = dados['Tipo'].isin(residencial)#atribuimos a filtragem à variável selecao
dados_residencial = dados[selecao]#utilizamos a variável para retirar do DataFrame dados os valores correspondentes a selecao, ou seja, os dados de imóveis residênciais
dados_residencial.head(10)
```

```python
list(dados_residencial['Tipo'].drop_duplicates())#criando uma lista com a column Tipo e tirando valores duplicados, verificamos se a filtragem funcionou
```

```python
dados_residencial.shape[0] #verificamos a quantidade de dados, ou seja, o número de rows após filtragem
```

```python
dados_residencial.index = range(dados_residencial.shape[0])#editamos o index para manter a sequência numérica
dados_residencial.head(10)
```

## Exportando a Base de Dados

```python
# Utilizamos to_csv para transformar DF em CSV e definimos o diretório em que queremos salvar
dados_residencial.to_csv('drive/MyDrive/Colab Notebooks/Formação Data Science/Pandas/Dados/aluguel_residencial.csv', sep = ';')
```
