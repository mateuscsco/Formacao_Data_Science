## Tratamento de Dados Faltantes 

```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive/')
```

```python
dados = pd.read_csv('drive/MyDrive/Colab Notebooks/Formação Data Science/Pandas/Dados/aluguel_residencial.csv', sep = ';')
dados.head(10)
```

```python
dados.isnull()#o método gera um DataFrame booleano, em que a observação marcada como True caracteriza um dado nulo, isto é, um missing
#temos também o método notnull() que funciona de maneira inversa
```

```python
dados.info()
# para verificarmos a quantidade de dados em cada column(quantidade de rows)
```

```python
dados['Valor'].isnull()#utilizamos o método isnull na column valor
```

```python
dados[dados['Valor'].isnull()]#exibimos dados, tal que o argumento retorne True, ou seja, rows em que Valor é nulo
```

```python
A = dados.shape[0]
dados.dropna(subset = ['Valor'], inplace = True)
B = dados.shape[0]
A - B
```

```python
#caso o imóvel seja um apartamento, os dados nulos serão excluídos.
dados[dados['Condominio'].isnull()].shape[0]#primeiro identificamos dados da column Condominio nulos
```

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
selecao#assinaturas que deverão ser removidas do banco de dados
```

```python
#Para eliminar as assinaturas nulas selecionadas ao invés de coletá-las, iremos adicionar ~ em selecao, que inverte a Series booleana.
A = dados.shape[0]
dados = dados[~selecao]
B = dados.shape[0]
A - B
```

```python
dados[dados['Condominio'].isnull()].shape[0]#verificar quantas assinaturas nulas temos em Condominio agora:
```

```python
dados.fillna(0, inplace = True)#manter esses dados, mas atribuir o valor 0 a eles
dados.info() 
#Temos, portanto, 21826 registros e nenhum deles é nulo
```

```python
dados = dados.fillna({'Condominio': 0, 'IPTU': 0})
#Outra maneira, é criar um dicionário que indicará os valores das variáveis. Podemos, inclusive, inserir valores diferentes para ambas as variáveis.
```

```python
dados.to_csv('drive/MyDrive/Colab Notebooks/Formação Data Science/Pandas/Dados/aluguel_residencial.csv', sep = ';', index = False)
```
