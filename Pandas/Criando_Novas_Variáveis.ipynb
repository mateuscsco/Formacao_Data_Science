##Criando Novas Variáveis

```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive/')
```

```python
dados = pd.read_csv('drive/MyDrive/Colab Notebooks/Formação Data Science/Pandas/Dados/aluguel_residencial.csv', sep = ';')
dados.head(10)
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
#criando nova variável (Valor Bruto) que recebe a soma do Valor, Condominio e IPTU
dados
```

```python
dados['Valor m2'] = dados['Valor']/dados['Area']
#criando variável valor m2  que recebe a divisão de Valor por Area
dados
```

```python
dados['Valor m2'] = dados['Valor m2'].round(2)
#para melhorar a visualização, utilizamos a função round(), que realiza arredondamento, dando o paramêtro 2 (duas casas decimais)
dados
```

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto']/dados['Area']).round(2)
#nova variável, referente ao valor bruto por area
dados
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
#criamos array com alguns tipos de imóveis
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
#criamos uma nova variável na DataFrame, utilizando o método apply()que aplica em cada termo da column, no caso, a função lambda
#a função lambda está comparando os termos da column Tipo com os termos na array 'casa', retornando Casa se corresponder a algum item da lista, senão, retornando Apartamento
dados
```

#Excluindo Variáveis

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
dados_aux.head(10)
```

```python
#Para excluirmos uma variável do banco de dados, usaremos o termo del, passaremos o DataFrame em questão e explicitaremos o nome da variável a ser deletada
del dados_aux['Valor Bruto']
dados_aux.head(10)
```

```python
#Outra maneira de excluir variáveis é utilizar o método pop(). Excluiremos neste exemplo a variável Valor Bruto m2:
dados_aux.pop('Valor Bruto m2')
```

```python
dados_aux.head(10)
```

```python
dados.drop(['Valor Bruto', 'Valor Bruto m2'], axis = 1, inplace = True)
#Uma maneira de excluirmos múltiplas variáveis é utilizando o método drop()
#precisaremos especificar o eixo em que se dará a exclusão, uma vez que ele pode atuar especificamente em linhas ou colunas. 
#Então, escreveremos axis = 1, afinal o eixo 0 corresponde às linhas e o 1 colunas. Precisaremos, ainda, assinalar inplace como True.
dados
```
