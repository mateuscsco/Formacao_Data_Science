```python
import pandas as pd
```

```python
data = [(1, 2, 3, 4),
        (5, 6, 7, 8),
        (8, 10, 11, 12),
        (13, 14, 15, 16)]
df = pd.DataFrame(data, 'l1 l2 l3 l4'.split(), 'c1 c2 c3 c4'.split())#data=data, 'l1 l2 l3 l4'.split()=index e 'c1 c2 c3 c4'.split()= columns
```

```python
'l1 l2 l3 l4'.split() # o método split() divide uma string em elementos individualizados
```

```python
df
```

```python
df['c1']#para selecionar uma column no formato de Series
```

```python
type(df['c1'])#provando que o resultado é do tipo Series
```

```python
df[['c3','c1']]#selecionando duas columns
```

```python
df[:]#selecionando a DataFrame toda
```

```python
df[1:3]#seleção somente das linhas l2e l3
```

```python
df[1:][['c3', 'c1']]#selecionando à partir da row(linha) 1 (dado que a contagem se inicia em 0), as columns c3 e c1
```

```python
df.loc['l3']#permite que realizemos seleções a partir do rótulo das linhas, em forma de Series
```

```python
df.loc[['l3', 'l2']]#selecionando mais de uma row
```

```python
df.loc['l1', 'c2']#podemos selecionar uma row e uma column, como uma notação matricial
```

```python
df.iloc[0,1]#utiliza os índices numéricos. Se quisermos coletar l1, precisaremos escrever 0; para coletar a coluna c1, escreveremos 1
```

```python
df.loc[['l3', 'l1'], ['c4','c1']]#para selecionar um grupo de rows e columns específicos
```

```python
df.iloc[[2,0],[3,0]]#a mesma seleção com o iloc[]
```
