```python
%matplotlib inline #configuração para plotagem dos gráficos
import pandas as pd
from google.colab import drive
drive.mount('/content/drive/')
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15,8))
```

```python
dados = pd.read_csv('drive/MyDrive/Colab Notebooks/Formação Data Science/Pandas/Extras/dados/aluguel.csv', sep = ';')
dados.head()
```

```python
#Criaremos a área que sustentará o gráfico:
area = plt.figure() 
```

```python
#Iremos inserir quatro gráficos dentro desse área
g1 = area.add_subplot(2, 2, 1)#dentro dessa figura há um gráfico, duas linhas e duas colunas na posição 1
g2 = area.add_subplot(2, 2, 2)#dentro dessa figura há um gráfico, duas linhas e duas colunas na posição 2
g3 = area.add_subplot(2, 2, 3)#dentro dessa figura há um gráfico, duas linhas e duas colunas na posição 3
g4 = area.add_subplot(2, 2, 4)#dentro dessa figura há um gráfico, duas linhas e duas colunas na posição 4
```

```python
g1.scatter(dados.Valor, dados.Area)# gráfico de dispersão
g1.set_title('Valor X Área')

g2.hist(dados.Valor)#histograma
g2.set_title('Histograma')

dados_g3 = dados.Valor.sample(100) #amostra aleatória dentro do próprio dataframe
dados_g3.index = range(dados_g3.shape[0])# é necessário formatar o index para plotar o gráfico de 0 a 99
g3.plot(dados_g3)
g3.set_title('Amostra (Valor)')

grupo = dados.groupby('Tipo')['Valor']
label = grupo.mean().index
valores = grupo.mean().values
g4.bar(label, valores)#gráfico de barras
g4.set_title('Valor Médio por Tipo')
area
```

```python
area.savefig('grafico.png', dpi = 300, bbox_inches = 'tight')#salvando a imagem
```
