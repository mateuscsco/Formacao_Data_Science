```python
import pandas as pd
```

```python
data = [1, 2, 3, 4, 5]#criamos uma array unidimensional
s = pd.Series(data)#tornamos a array uma Serie atribuida a s
s#visualizamos que, para a Serie é tb atribuída uma contagem de index
```

```python
index = ['Linha' + str(i) for i in range(5)]#realiza um contador, de 0 a 4, por meio do for, o str(i) tona o i uma string que é concatenada à 'Linha'
index
```

```python
s = pd.Series(data = data, index = index)#podemos passar o data e o index para a Series como "argumentos" para a criação da Series
s
```

```python
s1 = s + 2#podemos realizar operações matemáticas com a Serie, como no caso, somando 2 aos valores das rows
s1
```

```python
s2 = s + s1# podemos até mesmo realizar somas entre Series
s2
```

```python
#@title Array de arrays
data = [[1, 2, 3], 
        [4, 5, 6], 
        [7, 8, 9]] 
```

```python
df1 = pd.DataFrame(data=data)#criando um DataFrame
df1
```

```python
index = ['Linha' + str(i) for i in range(3)]
columns = ['Coluna' + str(i) for i in range(3)]
df1=pd.DataFrame(data = data, index = index, columns = columns)#utilizando argumentos de index e columns
df1
```

```python
columns = ['Coluna' + str(i) for i in range(3)]
df2 = pd.DataFrame(data = data, index = index, columns = columns)
df2
```

```python
data = {'Coluna0': {'Linha0': 1, 'Linha1': 4, 'Linha2': 7},#obtendo o mesmo DataFrame com dicionário
        'Coluna1': {'Linha0': 2, 'Linha1': 5, 'Linha2': 8},
        'Coluna2': {'Linha0': 3, 'Linha1': 6, 'Linha2': 9}}
df3 = pd.DataFrame(data)
df3
```

```python
data = [(1, 2, 3), # obtendo a mesma Data Frame com tuplas
        (4, 5, 6), 
        (7, 8, 9)]
```

```python
df4 = pd.DataFrame(data = data, index = index, columns = columns)
df4
```

```python
df1[df1 > 0] = 'A' #transformando termos maiores que 0 em A
df1
```

```python
df2[df2 > 0] = 'B' #transformando termos maiores que 0 em B
df2
```

```python
df3[df3 > 0] = 'C'#transformando termos maiores que 0 em C
df3
```

```python
df4 = pd.concat([df1, df2, df3]) # concatenando as DataFrames no axis = 0
df4
```

```python
df4 = pd.concat([df1, df2, df3], axis = 1) # concatenando as DataFrames no axis = 1
df4
```
