```python
import pandas as pd
```

```python
data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]#lista de listas
```

```python
list('321') #list que recebe string
```

```python
df = pd.DataFrame(data, list('321'), list('ZYX'))#data=data list=index e zyx como columns
df
```

```python
df.sort_index(inplace = True)#método sort organiza o data frame, no caso, por meio dos index
df
```

```python
df.sort_index(inplace = True, axis = 1)#informando o argumento axis, teremos uma nova lógica de organização, invertendo a ordem pelo label da linha e coluna.:
df
```

```python
df.sort_values(by = 'X', inplace = True)#Podemos, ainda, utilizar sort_values() e passar uma ou mais variáveis. Como exemplo inicial passaremos o apenas a variável X.O índice foi invertido, com base na organização da variável X
df
```

```python
df.sort_values(by = ['X','Y'], inplace = True)
df
```

```python
df.sort_values(by = '3', axis = 1, inplace = True)#Adicionaremos o termo axis = 1 e especificaremos a linha 3.
df
```
