```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
dados = pd.read_csv('drive/MyDrive/Colab Notebooks/Formação Data Science/Pandas/Dados/aluguel_residencial.csv', sep = ';')
dados.head(10)
```

```python
#para obter a média dos valores.
dados['Valor'].mean()
```

```python
#Nosso próximo passo agora é não apenas coletar a média geral, mas sim a média segundo alguns tipos de variáveis.
#Para facilitar a visualização, selecionaremos apenas alguns bairros
bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
selecao = dados['Bairro'].isin(bairros)
dados = dados[selecao]
```

```python
dados['Bairro'].drop_duplicates()#verificamos que realmente filtramos os bairros
```

```python
grupo_bairro = dados.groupby('Bairro')#criamos um dado do tipo DataFrameGroupBy a partir da column Bairro dados
type(grupo_bairro)#podemos verificar o tipo
```

```python
#Temos uma propriedade chamada groups, que consiste em um dicionário em que a chave é variável utilizada como indexador e os índices em que os registros de bairro foram encontrados.
#Dessa forma, é criada uma indexação do nosso DataFrame para facilitar consultas e crianção de tabelas.
for bairro, dados in grupo_bairro: 
    print(bairro)
```

```python
#Quando analisamos os dados por meio do método type(), verificaremos que temo um DataFrame para cada bairro. 
#Agora extrairemos o valor médio com base em nossa lista de bairros
for bairro, dados in grupo_bairro: 
    print('{} -> {}'.format(bairro, dados.Valor.mean()))
```

```python
grupo_bairro['Valor'].mean()
```

```python
grupo_bairro[['Valor', 'Condominio']].mean().round(2)
```

```python
#Teremos como resultado um DataFrame com as colunas:
#count a frequência; mean a média; std o desvio padrão; min o valor mínimo; 25% o primeiro quartio, 50% a mediana, 75% o terceiro quartio e max, o valor máximo.
grupo_bairro['Valor'].describe().round(2)
```

```python
#Essas informações estão errôneas, muito provavelmente esses valores correspondem à venda e não à locação.
#Para solucionar essa questão. Coletamos a variável Valor, e em seguida utilizaremos o método aggregate() que receberá como parâmetro o conjunto de estatísticas que nos interessam.
#Podemos, ainda, organizar melhor as nomenclaturas via rename()
grupo_bairro['Valor'].aggregate(['min', 'max', 'sum']).rename(columns = {'min': 'Mínimo', 'max': 'Máximo'})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt #importando a biblioteca
plt.rc('figure', figsize = (20,10)) #setando o tamanho da imagem
```

```python
fig = grupo_bairro['Valor'].std().plot.bar(color = 'blue') #setamos, cor, tipo de plot(bar) referente a função std() do Valor, ou seja, o desvio padrão
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color = 'blue')#Plotando agora o gráfico referente a média dos valores
fig.set_ylabel('Valor do Aluguel')#seta o label do valor y
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})#seta o título e seu tamanho da fonte
```
